```python
import pandas as pd
import numpy as np

# Copia del dataframe (buena práctica)
df2 = df.copy()

# Renombrar columnas clave para facilitar el análisis
df2 = df2.rename(columns={
    "¿Te enseñaron formalmente a usar IA en tu carrera?": "formacion_ia",
    "Puedo evaluar si una respuesta generada por IA es confiable?": "eval_confianza",
    "Sé cómo formular buenos prompts?": "buenos_prompts",
    "Creo que la IA puede afectar la integridad académica.": "riesgo_etico"
})

# Crear índices compuestos
df2["habilidad_tecnica"] = df2[["eval_confianza", "buenos_prompts"]].mean(axis=1)
df2["preocupacion_etica"] = df2["riesgo_etico"]

# Estandarizar categorías de formación
df2["formacion_ia"] = df2["formacion_ia"].astype(str).str.strip().str.title()
```

```python
# 3. Estadística descriptiva
```

```python
# 3.1 Frecuencias y porcentajes: formación (formal vs autodidacta)
tabla_formacion = (
    df2["formacion_ia"]
    .value_counts(dropna=False)
    .to_frame("Frecuencia")
)

tabla_formacion["Porcentaje"] = (
    df2["formacion_ia"]
    .value_counts(normalize=True, dropna=False) * 100
).round(2)

tabla_formacion
```

```python
tabla_formacion.to_excel("outputs/frecuencias_formacion_ia.xlsx")
```

```python
# 3.2 Descriptivos de variables clave
resumen_descriptivo = df2[
    ["habilidad_tecnica",
     "preocupacion_etica",
     "eval_confianza",
     "buenos_prompts",
     "riesgo_etico"]
].agg(["count", "mean", "std", "min", "max"]).T

resumen_descriptivo
```

```python
resumen_descriptivo.to_excel("outputs/descriptivos_variables_clave.xlsx")
```

```python
# 4. Correlación de Pearson
```

```python
# 4.1 Asegurar formato numérico
df2["habilidad_tecnica"] = pd.to_numeric(df2["habilidad_tecnica"], errors="coerce")
df2["preocupacion_etica"] = pd.to_numeric(df2["preocupacion_etica"], errors="coerce")

# 4.2 Correlación de Pearson (r)
r = df2["habilidad_tecnica"].corr(df2["preocupacion_etica"], method="pearson")
r
```

```python
pd.DataFrame({"r_pearson": [r]}).to_excel("outputs/correlacion_pearson.xlsx", index=False)
```

```python
# 5. Análisis comparativo
```

```python
# 5. Análisis comparativo
# Objetivo: comparar estudiantes con formación formal vs autodidacta y describir diferencias en uso de IA y promedios.

# Variable de comparación (grupos)
col_grupo = "formacion_ia"

# Variable categórica para tabla de contingencia
col_frecuencia_uso = "¿Con qué frecuencia usás IA (ChatGPT, Gemini, Copilot)?"
```

```python
# 5.1 Tabla de contingencia (frecuencias absolutas)
# Muestra cuántos estudiantes de cada grupo (formal/autodidacta) reportan cada nivel de frecuencia de uso de IA.

ct = pd.crosstab(
    df2[col_grupo],
    df2[col_frecuencia_uso],
    dropna=False,
    margins=True  # agrega totales por fila y columna
)

ct
```

```python
# Guardar tabla de contingencia (frecuencias absolutas) en Excel
ct.to_excel("outputs/contingencia_formacion_vs_frecuencia_ia.xlsx")
```

```python
# 5.2 Tabla de contingencia (porcentajes por fila)
# Normaliza por fila (por grupo) para ver la distribución porcentual interna de cada grupo.
# Ejemplo: dentro del grupo "Sí", qué % usa IA "Siempre", "A veces", etc.

ct_pct = pd.crosstab(
    df2[col_grupo],
    df2[col_frecuencia_uso],
    normalize="index",  # porcentajes por fila
    dropna=False
) * 100

ct_pct.round(2)
```

```python
# Guardar tabla de contingencia (porcentajes por fila) en Excel
ct_pct.round(2).to_excel("outputs/contingencia_porcentaje_fila.xlsx")
```

```python
# 5.3 Comparación de medias por grupo
# Calcula: tamaño de muestra (count), media (mean) y desviación estándar (std)
# para habilidad técnica y preocupación ética, diferenciando formación formal vs autodidacta.

cols_metricas = ["habilidad_tecnica", "preocupacion_etica"]

comparativo = (
    df2
    .groupby(col_grupo)[cols_metricas]
    .agg(["count", "mean", "std"])
)

comparativo
```

```python
# Guardar comparación de medias por formación en Excel
comparativo.to_excel("outputs/comparativo_medias_por_formacion.xlsx")
```
